```python
import stanza
from collections import Counter

# Initialize the StanfordNLP pipeline
with open('../Data/Alice/Alice.txt') as textFile:
    content = textFile.read()


# Process the text
nlp = stanza.Pipeline('en', processors='pos,tokenize,lemma')
doc = nlp(content)

for sentence in doc.sentences:
    for word in sentence.words:
        # print(word)
        if word.text == 'of':
            prevId = word.id - 1
            afterId = word.id + 1
            prevWord = next(
                (word for word in sentence.words if word.id == prevId), None)
            afterWord = next(
                (word for word in sentence.words if word.id == afterId), None)
            if (prevWord.upos == 'NOUN' and afterWord.upos == "NOUN"):
                print(f"{prevWord.text} of {afterWord.text}")
```
