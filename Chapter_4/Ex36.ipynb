```python
import json
from altair import value
import numpy as np
import matplotlib.pyplot as plt


with open('../Data/Output/Chapter4/Ex35-result.json', mode='r', encoding='utf-8') as f:
    stringContent = f.read()

words_count = json.loads(stringContent)

words_count_sort = sorted(
    words_count.items(), key=lambda x: x[1], reverse=True)

# print(words_count)
# print(words_count_sort)
top_10_list = words_count_sort[:10]
print(top_10_list)

left = np.arange(10)
heights = [item[1] for item in top_10_list]
labels = [item[0] for item in top_10_list]

plt.bar( left, heights, tick_label=labels)

# print(words_count.items())
# top_10_list = sorted(word_counts.items(), key=lambda x: x[1])
# print(top_10_list)
```
