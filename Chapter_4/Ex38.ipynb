```python
import matplotlib.pyplot as plt
import json
with open('../Data/Output/Chapter4/Ex35-result.json', mode='r', encoding='utf-8') as f:
    stringContent = f.read()

words_count = json.loads(stringContent)
words_count_sort = sorted(
    words_count.items(), key=lambda x: x[1])
heights = [item[1] for item in words_count_sort]
print(heights)
# max_height = max(heights)
# print(max_height)

plt.hist(x=heights, bins=100)
plt.savefig('../Data/Output/Chapter4/Ex38-histogram.png')
```
