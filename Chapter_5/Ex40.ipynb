```python
import json

from sympy import content


class Word:
    def __init__(self, text, lemma, pos):
        self.text = text
        self.lemma = lemma
        self.pos = pos


def extract_sentence_words(data):
    sentences = data["sentences"]
    if len(sentences) > 1:
        # Assuming the first sentence is the title, we skip it and consider the second sentence (index 1) as the body.
        body_sentence = sentences[0]
        words = body_sentence["tokens"]
        sentence = [Word(word["word"], word["lemma"], word["pos"])
                    for word in words]
        return sentence
    else:
        return []


with open("../Data/Chapter5/ai.en.txt.json", encoding='utf-8', mode='r') as f:
    strContent = f.read()

data = json.loads(strContent)
sentence_words = extract_sentence_words(data)

for word in sentence_words:
    print(f"Text: {word.text}\tLemma: {word.lemma}\tPOS: { word.pos}")
```
