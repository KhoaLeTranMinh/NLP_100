```python
import json



class Word:
    def __init__(self,sentence_index, index, text, lemma, pos, head = None, dep = None, children = None):
        self.sentence_index = sentence_index
        self.index = index
        self.text = text
        self.lemma = lemma
        self.pos = pos
        self.head = head
        self.dep = dep
        self.children = children


def extract_sentence_words(data):
    sentences = data["sentences"]
    result_sentences = []
    for sentence in sentences:
        words = sentence["tokens"]
        result_sentence = [Word(sentence_index = sentence["index"], index = word["index"], text = word["word"], lemma = word["lemma"], pos = word["pos"])
                    for word in words]
        
        dependencies = sentence["basicDependencies"]
        for word in result_sentence:
            word.children = []
            for dependency in dependencies:
                if dependency["dependent"] == word.index:
                    word.dep = dependency["dep"]
                    word.head = dependency
                if dependency["governor"] == word.index:
                    word.children.append(dependency)
        result_sentences.append(result_sentence)
    return result_sentences


with open("../Data/Chapter5/ai.en.txt.json", encoding='utf-8', mode='r') as f:
    strContent = f.read()

data = json.loads(strContent)
sentences = extract_sentence_words(data)

for sentence in sentences:
    for word in sentence:
        if word.dep == "ROOT":
            print(f'root word for sentence {word.sentence_index} is {word.text}') 
```
