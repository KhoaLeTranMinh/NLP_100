```python
import json



class Word:
    def __init__(self, index, text, lemma, pos, head = None, dep = None, children = None):
        self.index = index
        self.text = text
        self.lemma = lemma
        self.pos = pos
        self.head = head
        self.dep = dep
        self.children = children


def extract_sentence_words(data):
    sentences = data["sentences"]
    result_sentences = []
    for sentence in sentences:
        words = sentence["tokens"]
        result_sentence = [Word(word["index"], word["word"], word["lemma"], word["pos"])
                    for word in words]
        
        dependencies = sentence["basicDependencies"]
        for word in result_sentence:
            word.children = []
            for dependency in dependencies:
                if dependency["dependent"] == word.index:
                    word.dep = dependency["dep"]
                    word.head = dependency
                if dependency["governor"] == word.index:
                    word.children.append(dependency)
        result_sentences.append(result_sentence)
    return result_sentences


with open("../Data/Chapter5/ai.en.txt.json", encoding='utf-8', mode='r') as f:
    strContent = f.read()

data = json.loads(strContent)
sentences = extract_sentence_words(data)
first_sentence = sentences[0]
for word in first_sentence:
    print(f'Index:{word.index} \t Text: {word.text}\tLemma: {word.lemma}\tPOS: \t{ word.pos}\t dep:{word.dep}\t \thead:{word.head} \nchildren:{word.children}')
    
for word in first_sentence:
    for child in word.children:
        print(f'head:{child["governorGloss"]} \t dependent:{child["dependentGloss"]} \t dep:{child["dep"]}')
    
```
